```python
import numpy as np
import pandas as pd 
```

```python
dataset_movies = pd.read_csv('my_movies.csv',sep=',')
dataset_movies = dataset_movies.replace (np.nan,'',regex=True)
dataset_movies
```

```python
dataset_movies
transactions =[]
for i in range (0,9):
    transactions.append([str(dataset_movies.values[i,u])for u in range(1,15)])
```

```python
movies_series = pd.DataFrame(pd.Series(transactions))
movies_series
```

```python
movies_series.columns=['business']
movies_series
```

```python
X = movies_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(X,min_support=0.005,max_len=3,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending=False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rules.head(10)
rules.sort_values('lift',ascending = False).head(10)
```
