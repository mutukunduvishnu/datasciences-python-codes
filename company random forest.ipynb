```python
com_data_df = pd.DataFrame(com_data)
```

```python
# create the dummies 
dummies = pd.get_dummies(com_data_df['ShelveLoc']).rename(columns=lambda x: 'ShelveLoc_' + str())
# bring the dummies into the original dataset
com_data_df = pd.concat([com_data_df,dummies],axis=1)
print(com_data_df)
```

```python
com_data_df.head()
```

```python
com_data_df=com_data_df.drop('ShelveLoc',1)
```

```python
# create the dummies 
dummies = pd.get_dummies(com_data_df['Urban']).rename(columns=lambda x: 'Urban_' + str())
# bring the dummies into the original dataset
com_data_df = pd.concat([com_data_df,dummies],axis=1)
print(com_data_df)
```

```python
com_data_df= com_data_df.drop('Urban',1)
```

```python
# create the dummies 
dummies = pd.get_dummies(com_data_df['US']).rename(columns=lambda x: 'US_' + str())
# bring the dummies into the original dataset
com_data_df = pd.concat([com_data_df,dummies],axis=1)
print(com_data_df)
```

```python
com_data_df = com_data_df.drop('US',1)
```

```python
com_data_df.head()
```

```python
row_indexes=com_data_df[com_data_df['Sales']>=10].index
```

```python
com_data_df.loc[row_indexes,'sales']='yes'
```

```python
row_indexes=com_data_df[com_data_df['Sales']<10].index
```

```python
com_data_df.loc[row_indexes,'sales']='no'
```

```python
com_data_df.head()
```

```python
com_data_df = com_data_df.drop('sales',1)
```

```python
com_data_df.head()
```

```python
colnames = list(com_data_df.columns)
colnames
```

```python
predictors = colnames[:14]
target = colnames[14]
```

```python
target
```

```python
predictors
```

```python
x = com_data_df[predictors] 
y = com_data_df[target]
x
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=15,criterion='entropy')
```

```python
# fitting the random forest classification 
rf.fit(x,y) 
rf.estimators_
```

```python
rf.classes_  # class labels
rf.n_classes_ # number of labels in class level
```

```python
rf.n_features_  # Number of input features in model 14 here.
```

```python
rf.n_outputs_ # number of outputs when fit performed
```

```python
rf.oob_score_
```

```python
rf.predict(x)
```
